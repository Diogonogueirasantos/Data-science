```python
import pandas as pd
```

```python
df = pd.read_csv('social_Media_V1.csv')
```

```python
df.drop('Customer Name', axis=1, inplace=True)
```

```python
df.head()
```

```python
#changing the name of labels 

df.rename(columns={"Consumer ID":"id_Cliente", "Age":"Idade", "Income (USD)":"Ganho_Anual_(USD)", "Education Level":"Nivel_Escolar",
                   "Social Media Usage (Hours/Day)":"Tempo_Uso_Redes_Sociais", "Social Media Platforms":"Plataforma_de_Uso", "Influence Level":"Grau_de_Influencia",
                   "Purchase Decision":"Decisao_de_Compra", "Product Category":"Categoria_do_Produto", "Specific Product":"Produto", "Amount Spent (USD)":"Total_Gasto (USD)",
                   "Brand Name":"Marca", "City":"Cidade"}, inplace=True)
```

```python
df.rename(columns={"Gender":"Genero"}, inplace=True)
```

```python
df.loc[df['Nivel_Escolar'] == "High School"]
```

```python
df.head(1)
```

```python
#changing the values of variables
df['Genero'] = df['Genero'].replace({'Male':'Masculino', 'Female':'Feminino', 'Non-binary':'Nao-Binario'})
```

```python
df['Nivel_Escolar'] = df['Nivel_Escolar'].replace({"High School":"Ensino_Medio", "Bachelor's":"Graduacao", "Master's":"Mestrado", "PhD":"Doutorado"})
```

```python
df['Grau_de_Influencia'] = df['Grau_de_Influencia'].replace({"Somewhat Influential":"Alguma_Influencia", "Very Influential":"Muito_influenciavel", "Not Influential":"Nao_Influenciavel"})
```

```python
df['Decisao_de_Compra'] = df['Decisao_de_Compra'].replace({"Yes":"Sim", "No":"Nao"})
```

```python
df.head(5)
```

```python
df['Ganho_Anual_(USD)'] = df['Ganho_Anual_(USD)'].astype(int)
```

```python
df['Ganho_Mensal'] = df['Ganho_Anual_(USD)'] / 12
```

```python
df['Ganho_Mensal'] = df['Ganho_Mensal'].round()
df['Ganho_Mensal'] = df['Ganho_Mensal'].astype(int)
```

```python
df.head()
```

```python
df['Produto'] = df['Produto'].fillna('Nao convertido')
```

```python
df['Categoria_do_Produto'] = df['Categoria_do_Produto'].fillna('Indefinido')
```

```python
df['Marca'] = df['Marca'].fillna('Indefinido')
```

```python
titulos = ['id_Cliente', 'Idade', 'Genero', 'Ganho_Anual_(USD)', 'Ganho_Mensal',
           'Nivel_Escolar', 'Tempo_Uso_Redes_Sociais', 'Plataforma_de_Uso',
           'Grau_de_Influencia', 'Decisao_de_Compra', 'Categoria_do_Produto',
           'Produto', 'Total_Gasto (USD)', 'Marca', 'Cidade']
```

```python
df = df[titulos]
```

```python
df.head()
```

```python
df.to_csv('V2_SocialMedia.csv', index=False)
```
