```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('social_Media_V1.csv')
```

```python
df.head(1)
```

```python
df.rename(columns={'Consumer ID':'ID_Cliente', 'Customer Name':'Nome', 'Age':'Idade',
                   'Gender':'Genero', 'Education Level':'Nivel_de_educacao', 'Income (USD)':'Salario', 'Social Media Usage (Hours/Day)':'Tempo_de_Uso (Horas/Dias)',
                   'Social Media Platforms':'Redes_Sociais', 'Influence Level':'Nivel_de_Influencia', 'Purchase Decision':'Decisao_de_Compra',
                   'Product Category':'Categoria_do_Produto', 'Specific Product':'Produto', 'Amount Spent (USD)':'Total_Gasto (USD)', 'Brand Name':'Nome_da_Marca',
                   'City':'Cidade'}, inplace=True)
```

```python
df.head(5)
```

```python
df['Genero'] = df['Genero'].replace({"Male":"Masculino", "Female":"Feminino"})
```

```python
df.head(5)
```

```python
df.head(5)
```

```python
df['Renda_Mensal'] = df['Salario'].round(2) /12
```

```python
df.rename(columns={'Salario':'Ganho_Anual (USD)'}, inplace=True)
```

```python
df['Renda_Mensal'] = df['Renda_Mensal'].round(2)
```

```python
df.head(5)
```

```python
df.shape
```

```python
ordem_colunas = ['ID_Cliente', 'Nome', 'Idade', 'Genero', 'Ganho_Anual (USD)', 'Renda_Mensal', 'Nivel_de_educacao', 'Tempo_de_Uso (Horas/Dias)',
                 'Redes_Sociais', 'Nivel_de_Influencia', 'Decisao_de_Compra', 'Categoria_do_Produto', 'Produto', 'Total_Gasto (USD)', 
                 'Nome_da_Marca', 'Cidade']
```

```python
df = df[ordem_colunas]
```

```python
df
```

```python
df['Nivel_de_Influencia'] = df['Nivel_de_Influencia'].replace({'Very Influential':'Muito Influenciavel', 
                                                               'Somewhat Influential':'Alguma Influencia',
                                                               'Not Influential':'Nao Influenciavel'})
```

```python
df.head(1)
```

```python
df['Decisao_de_Compra'] = df['Decisao_de_Compra'].replace({'Yes':'Sim', 'No':'Nao'})
```

```python
df.head(1)
```

```python
stats.zscore(df['Renda_Mensal'])
```

```python
df['Ganho_Anual (USD)'] = df['Ganho_Anual (USD)'].astype(int)
```

```python
df = df.drop(columns=['Nome'])
```
