```python
import pandas as pd 
```

```python
df = pd.read_csv('Clientes_VI.csv')
```

```python
df.head(5)
```

```python
df['estado'] = df['estado'].str.upper()
```

```python
df['CPF'] = df['cpf'].apply(lambda cpf: f'{cpf[:3]}.***.***.-{cpf[-2:]}')
```

```python
df.drop('cpf', axis=1, inplace=True)
```

```python
df['data'] = pd.to_datetime(df['data'], dayfirst=True, errors='coerce')
```

```python
df['estado'] = df['estado'].str.upper()
```

```python
df.rename(columns={'estado': 'Cidade'}, inplace=True)
```

```python
df['Estado'] = df['endereco'].apply(lambda x: x.split('/')[-1]) 
```

```python
df['CEP'] = df['endereco'].apply(lambda x: x.split('\n')[2::1])
```

```python
df['endereco'] = df['endereco'].apply(lambda x: x.split('\n')[:])
```

```python
df.rename(columns={'cidade': 'CIDADE', 'estado': 'ESTADO'}, inplace=True)
```

```python
reorganize = ['NOME', 'DATA', 'IDADE', 'CPF', 'Cidade', 'Estado', 'ENDERECO', 'CEP']
```

```python
df = df[reorganize]
```

```python
df.to_csv('Clientes_final_version.csv', index=False)
```
