```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/performance_estudantil.csv')
```

```python
df.shape
```

```python
df
```

```python
df['Sexo']  = df['Sexo'].apply(lambda Sexo: 'Feminino' if Sexo=='Female' else Sexo)
```

```python
df['Sexo'] = df['Sexo'].replace({'Male':'Masculino'})
```

```python
df
```

```python
df['Atividades_Extracurriculares'] = df['Atividades_Extracurriculares'].replace({'No': 'Não', 'Yes':'Sim'})
```

```python
df
```

```python
df['Acesso_a_Internet'] = df['Acesso_a_Internet'].replace({'No':'Não', 'Yes':'Sim'})
```

```python
df['Nivel_Renda_Familiar'] = df['Nivel_Renda_Familiar'].replace({'Medium':'Medio', 'Low':'Baixo', 'High':'Alto'})
```

```python
df
```

```python
df['Nivel_Educao_dos_Pais'] = df['Nivel_Educao_dos_Pais'].replace({'High School':'Ensino_Medio', "Bachelor's":"Bacharelado", "Master's":"Mestrado", 'PhD':'Doutorado'})
```

```python
df['Mascara_Email'] = df['Email'].apply(lambda Email: f'{Email[:3]}*********************{Email[-3:]}')
```

```python
df['Email']  = df['Mascara_Email']
```

```python
df
```

```python
df.drop('Mascara_Email', axis=1, inplace=True)
```

```python
df
```

```python
df['Id_Estudante'] = df['Id_Estudante'].apply(lambda Id_Estudante: f'{Id_Estudante[:1]}***{Id_Estudante[-1:]}')
```

```python
df
```

```python
df['Curso'] = df['Curso'].replace({'Business':'Gestao', 'Mathematics':'Exatas', 'CS':'Ciencia da Computacao'})
```

```python
df['Media'] = df['Media'].round()
```

```python
df['Frequencia'] = df['Frequencia'].round()
```

```python
df['Nota_Final'] = df['Nota_Final'].round()
```

```python
df['Pontos_por_projetos'] = df['Pontos_por_projetos'].round()
```

```python
df['Pontos_por_Participacao'] = df['Pontos_por_Participacao'].round()
```

```python
df['Media_por_Atribuicoes'] = df['Media_por_Atribuicoes'].round()
```

```python
df['Media_por_Testes'] = df['Media_por_Testes'].round()
```

```python
df['Pontos_Totais'] = df['Pontos_Totais'].round()
```

```python
df.to_csv('Performance_Estudantil.csv', index=False)
```
