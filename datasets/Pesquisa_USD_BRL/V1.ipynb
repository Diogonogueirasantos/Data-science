```python
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv('USD_BRL_hist.csv')
```

```python
df['Data'] = pd.to_datetime(df['Data'], dayfirst=True, errors='coerce')
```

```python
df['Dia'] = df['Data'].map(lambda x: x.day)
```

```python
df['Mes'] = df['Data'].map(lambda x: x.month)
```

```python
df['Ano'] = df['Data'].map(lambda x: x.year)
```

```python
df['USD_BRL'] = df['USD_BRL'].round(2)
```

```python
df.head()
```

```python
df.to_csv('USD_BRL_Hist_V1.csv', index=False)
```

```python
sns.lineplot(x=df['Ano'], y=df['USD_BRL'], color='#023535')
```

```python
sns.histplot(df, x=df['USD_BRL'], hue=df['Ano'], log_scale=True)
```

```python
sns.set_theme(style='whitegrid')
sns.scatterplot(x=df['Ano'], y=df['USD_BRL'], color='#023535')
```
