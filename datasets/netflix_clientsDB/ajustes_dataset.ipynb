```python
#Alterando o nome dos rótulos
df.rename(columns={"User_ID":"ID_Usuário", "Name":"Nome","Country":"País", "Age":"Idade", "Subscription_Type":"Assinaturas",
                   "Watch_Time_Hours":"Tempo_assistido", "Favorite_Genre":"Genero_favorito", "Last_Login":"Ultimo_Login"}, inplace=True)
```

```python
df.head(10)
```

```python
#Alterando o nome das categorias 
df["Assinaturas"].replace({"Basic":"Basico", "Standard":"Padrao"}, inplace=True)
```

```python
df
```

```python
df.drop(axis=0, columns="Nome", inplace=True)
```

```python
df
```

```python
#aplicação a padronização de datas para o campo Ultimo Login
df['Ultimo_Login'] = pd.to_datetime(df["Ultimo_Login"], yearfirst=True, errors="coerce")
```

```python
df.dtypes
```

```python
df["Genero_favorito"]
```

```python
#alterando as categorias para a coluna genero favorito 
df["Genero_favorito"].replace({"Action":"Acao", "Documentary":"Documentario", "Sci-Fi":"Ficcao cientifica", "Comedy":"Comedia"}, regex=True, inplace=True)
```

```python
df["Genero_favorito"]
```

```python
#Salvando dataset modificado

df = df.to_csv('netflix_user_v1.csv', index=False)
```
